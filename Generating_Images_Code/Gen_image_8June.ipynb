```python
filenames = [img for img in glob.glob("*.csv")]
```

```python
import re
tokenize = re.compile(r'(\d+)|(\D+)').findall
def natural_sortkey(string):          
    return tuple(int(num) if num else alpha for num, alpha in tokenize(string))

files = sorted(filenames, key=natural_sortkey)

files
```

```python
import os
number = 0
for exc in files:
    number = number + 1
    print("CSV file = ",exc)
    data = np.genfromtxt(exc,delimiter=',',dtype=None)
    d = np.zeros(((len(data[:,2])),(len(data[:,2]))),dtype = float)
    #Distance matrix
    d = np.zeros(((len(data[:,2])-1),(len(data[:,2])-1)),dtype = float)
    for i in range(1,len(data[1:,1])):
        for j in range(i,len(data[:,1])):
            d[i-1][j-1] = ((float(data[j,0])-float(data[i,0]))**2+(float(data[j,1])-float(data[i,1]))**2+(float(data[j,2])-float(data[i,2]))**2)**0.5
            #print("d{}{}".format(i,j),d[i][j])
    p = np.amax(d)



    print(p)
    q = 0
    r = 0
    for q in range(len(data[:,2])-1):
        for r in range(len(data[:,2])-1):
            k = 256/p
            d[q][r] = k*d[q][r]
    b = np.zeros((((len(data[:,2])-1))),dtype = float)
    n = np.zeros((((len(data[:,2])-1))),dtype = float)
    t = np.zeros((((len(data[:,2])-1))),dtype = float)
 
    for i in range(len(data[1:,3])):
        b[i] = float(data[1+i,3])
        n[i] = float(data[1+i,4])
        t[i] = float(data[1+i,6])
#Normalising b_type
    i = 0
    q = np.amax(b)

    for i in range(len(b)):
        k = 256/q
        b[i] = k*b[i]

#Normalising n_type
    i = 0
    q = np.amax(n)
    for i in range(len(n)):
        k = 256/q
        n[i] = k*n[i]


#Normalising Theta
    i=0
    k = 256/360
    for i in range(len(t)):
        #k = 256/360
        t[i] = k*t[i]
#Adding Theta Values in d matrix
    i = 1
    for i in range(len(t)):
        if i%2 != 0 :
            d[i][i-1] = t[i-1]
        else:
            continue
#Adding the values of b type and n-type in d matrix
    for k in range(len(b)):
        if (k%2) == 0:
            d[k][k]=b[k]
        else :
            d[k][k] = n[k]

    #plotting the result 
    print("Image generated = ", "36D_1_{}.png".format(number))
    fig , ax = plt.subplots(figsize = (4,4))
    plt.imshow(d,interpolation = "nearest",cmap = "Greys",)
    plt.axis('off')   
    plt.savefig("../36D_1_Images/36D_1_{}.png".format(number),bbox_inches='tight' ) 
    plt.show()
```
