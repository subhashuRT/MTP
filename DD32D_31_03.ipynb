```python
a1 = data[data['Strain'] == y]["de1"].loc[z]
a2 = data[data['Strain'] == y]["de2"].loc[z]
a3 = data[data['Strain'] == y]["de3"].loc[z]
a4 = data[data['Strain'] == y]["de4"].loc[z]
a5 = data[data['Strain'] == y]["de5"].loc[z]
a6 = data[data['Strain'] == y]["de6"].loc[z]
a7 = data[data['Strain'] == y]["de7"].loc[z]
a8 = data[data['Strain'] == y]["de8"].loc[z]
a9 = data[data['Strain'] == y]["de9"].loc[z]
a10 = data[data['Strain'] == y]["de10"].loc[z]
a11 = data[data['Strain'] == y]["de11"].loc[z]
a12 = data[data['Strain'] == y]["de12"].loc[z]
```

```python
l = [a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12]
for i in range(len(l)):
    #print(a[i])
    l[i] = (l[i])/100000000000
print(l)

Nu_slip = np.arange(1,13)
```

```python
fig = plt.figure(figsize = (10,10))
ax = plt.axes()
ax.bar(Nu_slip,l)
mean_slip = [np.mean(l)]*len(Nu_slip)
std_slip = [np.std(l)]*len(Nu_slip)
mean_line = ax.plot(Nu_slip,mean_slip, label='Mean={}'.format('%.2E' % Decimal(mean_slip[0])), linestyle='--')
std_dev = ax.plot(Nu_slip,std_slip, label = "Standard Deviation ={}".format('%.2E' % Decimal(std_slip[0])), linestyle="-.")
legend = ax.legend(loc='upper right')
plt.title("Dislocation Density on different slip system for {}% strain in 32D".format(y*100),fontsize = 20, y=1.01)
plt.xlabel("Slip sytem",fontsize = 20)
plt.ylabel("Dislocation Density (X {} {})".format(r'$10^{11}$',r'$m^{-2}$'),fontsize = 20)

plt.xticks(fontsize=20)
plt.yticks(fontsize = 20)
plt.savefig("DD({}).png".format(y*100))
plt.show()
```
